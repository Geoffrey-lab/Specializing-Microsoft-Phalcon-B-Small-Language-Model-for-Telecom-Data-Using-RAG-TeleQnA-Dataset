```python
import requests
import json
import csv
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm

# Function to generate an answer using the RAG pipeline
def generate_answer_rag(question):
    url = 'https://a7tncd274b.execute-api.eu-west-3.amazonaws.com/InitialStage/Falcon7B'
    data = {
        "inputs": question,
        "parameters": {
            "max_new_tokens": 50,
            "return_full_text": False,
            "do_sample": True,
            "top_k": 10
        }
    }
    json_data = json.dumps(data)
    headers = {"Content-Type": "application/json"}
    try:
        response = requests.post(url, data=json_data, headers=headers)
        if response.status_code == 200:
            response_data = response.json()
            body = response_data.get('body')
            if body:
                body_data = json.loads(body)
                if body_data and isinstance(body_data, list) and 'generated_text' in body_data[0]:
                    return body_data[0]['generated_text'].strip()
                else:
                    print("Unexpected body format:", body_data)
                    return None
            else:
                print("No body in response:", response_data)
                return None
        else:
            print("Failed to get a response:", response.status_code, response.text)
            return None
    except Exception as e:
        print("An error occurred:", e)
        return None

# Load training and testing data
def load_data(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

training_data = load_data('TeleQnA_training.txt')
testing_data = load_data('TeleQnA_testing1.txt')
questions_new = load_data('questions_new.txt')

# Initialize TF-IDF Vectorizer with stopwords removal
vectorizer = TfidfVectorizer(stop_words='english')

# Prepare submission data
submission_data = []

# Process training data
for question_id, item in tqdm(training_data.items(), desc="Processing Training Questions", unit="question"):
    question = item['question']
    answer_id = item['answer']  # Directly use the provided answer
    submission_data.append([question_id.replace('question ', ''), answer_id, 'Falcon 7.5B'])

# Process testing data
for question_id, item in tqdm(testing_data.items(), desc="Processing Testing Questions", unit="question"):
    question = item['question']
    generated_answer = generate_answer_rag(question)
    
    # Prepare options and filter out None values
    options = [item[f'option {i}'] for i in range(1, 6) if f'option {i}' in item and item[f'option {i}'] is not None]
    if not options:
        print(f"No options found for question {question_id}")
        answer_id = "0"  # Default to an invalid option if no options found
    else:
        options.append(generated_answer if generated_answer else "")
        vectors = vectorizer.fit_transform(options)
        similarity_matrix = cosine_similarity(vectors)
        generated_index = len(options) - 1
        similarities = similarity_matrix[generated_index][:-1]
        closest_option_index = similarities.argmax() + 1
        
        # Ensure we select an option based on the slightest similarity
        if similarities.max() <= 0:
            closest_option_index = similarities.argmin() + 1
            print(f"Answer not found for question {question_id}, selecting least dissimilar option: {closest_option_index}")
        
        answer_id = closest_option_index

    submission_data.append([question_id.replace('question ', ''), answer_id, 'Falcon 7.5B'])

# Process new questions
for question_id, item in tqdm(questions_new.items(), desc="Processing New Questions", unit="question"):
    question = item['question']
    generated_answer = generate_answer_rag(question)
    
    # Handle missing options and filter out None values
    options = [item.get(f'option {i}', "") for i in range(1, 6) if item.get(f'option {i}', "") is not None]
    options.append(generated_answer if generated_answer else "")
    vectors = vectorizer.fit_transform(options)
    similarity_matrix = cosine_similarity(vectors)
    generated_index = len(options) - 1
    similarities = similarity_matrix[generated_index][:-1]
    closest_option_index = similarities.argmax() + 1
    
    # Ensure we select an option based on the slightest similarity
    if similarities.max() <= 0:
        closest_option_index = similarities.argmin() + 1
        print(f"Answer not found for new question {question_id}, selecting least dissimilar option: {closest_option_index}")
    
    answer_id = closest_option_index

    submission_data.append([question_id.replace('question ', ''), answer_id, 'Falcon 7.5B'])

# Ensure all question IDs are included in the submission
all_question_ids = set(training_data.keys()).union(testing_data.keys()).union(questions_new.keys())
submitted_question_ids = set(item[0] for item in submission_data)
missing_ids = all_question_ids - submitted_question_ids

# Add missing IDs with default value
for missing_id in missing_ids:
    submission_data.append([missing_id.replace('question ', ''), "0", 'Falcon 7.5B'])

# Save to CSV
try:
    with open('submission.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Question_ID', 'Answer_ID', 'Task'])
        writer.writerows(submission_data)
    print("Submission file generated: submission.csv")
except PermissionError as e:
    print("Permission error occurred:", e)
    print("Please ensure the file is not open in any other program and you have the necessary permissions.")
```

```python

import csv

# Function to clean the Answer_ID column
def clean_answer_id(answer_id):
    # Extract the option number (the first word after "option")
    parts = answer_id.split(' ')
    if len(parts) > 1:
        return parts[1].replace(":", "")  # Remove any trailing colon
    return answer_id

# Function to clean the submission data
def clean_submission_data(input_file, output_file):
    with open(input_file, mode='r', newline='', encoding='utf-8') as infile, \
         open(output_file, mode='w', newline='', encoding='utf-8') as outfile:
        
        reader = csv.reader(infile)
        writer = csv.writer(outfile)
        
        # Write the header
        header = next(reader)
        writer.writerow(header)
        
        # Process each row
        for row in reader:
            question_id, answer_id, task = row
            cleaned_answer_id = clean_answer_id(answer_id)
            writer.writerow([question_id, cleaned_answer_id, task])

# Path to the input and output files
input_csv_file = 'submission.csv'
output_csv_file = 'cleaned_submission.csv'

# Clean the submission data
clean_submission_data(input_csv_file, output_csv_file)

print("Cleaned submission file generated:", output_csv_file)

#This code reads the `submission.csv` file, cleans the `Answer_ID` column to keep only the numerical value, and writes the cleaned data to a new CSV file called `cleaned_submission.csv`.
```
